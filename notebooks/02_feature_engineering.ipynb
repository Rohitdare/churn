```python
import pandas as pd
from datetime import datetime
```

```python
accounts = pd.read_csv("../data/raw/ravenstack_accounts.csv")
```

```python
accounts = accounts.drop(columns=["account_name"])
```

```python
accounts["signup_date"] = pd.to_datetime(accounts["signup_date"])
```

```python
today = pd.Timestamp(datetime.today().date())
accounts["account_age_days"] = (today - accounts["signup_date"]).dt.days
```

```python
accounts = accounts.drop(columns=["signup_date"])
```

```python
accounts["is_trial"] = accounts["is_trial"].astype(int)
accounts["churn_flag"] = accounts["churn_flag"].astype(int)
```

```python
categorical_cols = ["industry", "country", "referral_source", "plan_tier"]

accounts_encoded = pd.get_dummies(
    accounts,
    columns=categorical_cols,
    drop_first=True
)
```
